    *   Selecione 'JSON' como o tipo de chave e clique em 'Criar'.

6.  **Compartilhe sua Planilha Google com a Conta de Serviço:**

## Integração e Qualificação com LLM (Atualizada)

### Subtask:
Atualizar o `system prompt` do LLM (Gemini ou ChatGPT) para extrair, além das informações anteriores, os detalhes necessários para o cálculo de preço (e.g., `quantidade` ou `medidas` mais precisas se aplicável) e para o agendamento de visitas (`solicita_visita`, `data_preferencial_visita`, `endereco_visita`), retornando-os como um objeto JSON. O `status_pedido` também deve ser mais dinâmico (e.g., 'QUALIFICADO', 'ORÇADO', 'VISITA AGENDADA').


**Reasoning**:
The subtask requires defining a detailed `system_prompt` for the LLM, setting up the Gemini model with this prompt, and creating a function to call the LLM and parse its JSON output. I will start by defining the `system_prompt` based on the specified extraction requirements.



````python
import google.generativeai as genai
import json

# 1. Definir o system_prompt detalhado para o LLM
system_prompt = """
Você é um assistente de qualificação de pedidos de cortinas para a Tati Cortinas. Sua tarefa é extrair informações de mensagens de clientes e formatá-las em um objeto JSON. Você deve ser preciso e rigoroso na extração de dados.

Instruções:
- Extraia as seguintes informações da mensagem do cliente:
  - cliente_nome: Nome do cliente (string). Se não mencionado, tente inferir ou deixe como null.
  - cliente_telefone: Telefone de contato do cliente (string, formato livre). Se não mencionado, tente inferir ou deixe como null.
  - tipo_cortina: O tipo principal de cortina (string). Deve ser um dos seguintes, se aplicável: 'Voil', 'Blackout', 'Persiana'. Se não for claramente um desses, deixe como null.
  - subtipo_cortina: O subtipo da cortina (string). Exemplos: 'Cor_Unica', 'Estampado' para Voil; 'Tecido', 'PVC' para Blackout; 'Horizontal', 'Vertical' para Persiana. Se não for especificado ou aplicável, deixe como null.
  - material_cortina: O material específico para persianas (string). Exemplos: 'Aluminio', 'Madeira'. Se não for persiana ou não especificado, deixe como null.
  - largura: Medida da largura da janela/parede em metros (float). Se não especificado, deixe como null.
  - altura: Medida da altura da janela/parede em metros (float). Se não especificado, deixe como null.
  - quantidade: Número de unidades da cortina/persiana (integer). Se não especificado, assuma 1, se não for inferível deixe como null.
  - solicita_instalacao: Booleano (true/false) se o cliente explicitamente solicita instalação. Se não mencionado, considere false.
  - solicita_visita: Booleano (true/false) se o cliente solicita uma visita técnica para medição/orçamento. Se não mencionado, considere false.
  - data_preferencial_visita: Data ou período preferencial para a visita (string, formato livre). Se não mencionado, deixe como null.
  - endereco_visita: Endereço completo para a visita (string). Se 'solicita_visita' for true e o endereço não for fornecido, deixe como null e priorize solicitar o endereço.
  - condicao_pagamento: Se o cliente menciona 'à vista' ou outra condição de pagamento (string). Se não mencionado, deixe como null.
  - status_pedido: Classifique o status do pedido com base nas informações extraídas:
    - 'QUALIFICADO': Se todas as informações para orçamento (tipo_cortina, largura, altura, quantidade) estiverem presentes.
    - 'VISITA SOLICITADA': Se 'solicita_visita' for true, independentemente de outros detalhes.
    - 'INCOMPLETO': Se faltarem informações cruciais para orçamento ou visita, e nenhuma das condições acima for atendida.

- Retorne APENAS o objeto JSON com as chaves e valores extraídos. NÃO inclua nenhum texto adicional, explicações, saudações ou formatação além do JSON.
- Se a mensagem for irrelevante ou não contiver informações para qualificação de pedido, retorne um JSON vazio ou com 'status_pedido': 'INCOMPLETO'.
"""

# 4. Definir o modelo do Google Gemini e configurar com o system_prompt
# Certifique-se de que a API key já foi configurada em um passo anterior.
model = genai.GenerativeModel('gemini-pro', system_instruction=system_prompt)

print("System prompt definido e modelo Gemini configurado com sucesso.")


def qualify_order_with_llm(message):
    """
    Envia a mensagem do cliente ao LLM configurado com o system_prompt
    e retorna o objeto JSON extraído.
    """
    try:
        # Envia a mensagem ao LLM
        response = model.generate_content(message)

        # Verifica se a resposta contém texto e tenta parsear como JSON
        if response.text:
            # Limpar o texto para garantir que apenas o JSON seja parseado
            json_text = response.text.strip().replace('```json', '').replace('```', '')
            extracted_data = json.loads(json_text)
            return extracted_data
        else:
            print("LLM response text is empty.")
            return {'status_pedido': 'INCOMPLETO', 'observacao': 'Resposta vazia do LLM'}
    except json.JSONDecodeError as e:
        print(f"Erro ao decodificar JSON da resposta do LLM: {e}")
        print(f"Resposta bruta do LLM: {response.text}")
        return {'status_pedido': 'INCOMPLETO', 'observacao': f'Erro de JSON: {e}'}
    except Exception as e:
        print(f"Ocorreu um erro ao chamar o LLM: {e}")
        return {'status_pedido': 'INCOMPLETO', 'observacao': f'Erro do LLM: {e}'}
````

## Lógica de Cálculo de Preço

### Subtask:
Desenvolver uma função que, com base nas informações extraídas pelo LLM (como `tipo_cortina` e `quantidade`/`medidas`), calcule o `preco_total` do pedido utilizando a tabela de preços corrigida. O resultado deve ser incluído nos dados do pedido.


**Reasoning**:
The subtask requires developing a function to calculate the total price of an order based on LLM extracted details and a price table. I will define the `calculate_price` function according to the specified instructions, including logic for curtain types, dimensions, quantity, installation, and discounts.



````python
def calculate_price(order_details, price_table):
    """
    Calcula o preço total de um pedido de cortinas com base nas informações extraídas
    e na tabela de preços.

    Args:
        order_details (dict): Dicionário contendo as informações do pedido extraídas pelo LLM.
        price_table (dict): Dicionário contendo a tabela de preços.

    Returns:
        float: O preço total calculado do pedido.
    """
    total_price = 0.0

    # 3. Extrair informações relevantes
    tipo_cortina = order_details.get('tipo_cortina')
    subtipo_cortina = order_details.get('subtipo_cortina')
    material_cortina = order_details.get('material_cortina') # Usado para Persianas
    largura = order_details.get('largura')
    altura = order_details.get('altura')
    quantidade = order_details.get('quantidade', 1) # Assume 1 se não especificado
    solicita_instalacao = order_details.get('solicita_instalacao', False)
    condicao_pagamento = order_details.get('condicao_pagamento')

    # 4. Calcule a área em metros quadrados (m2)
    area_m2 = 0.0
    if largura is not None and altura is not None:
        area_m2 = largura * altura
    else:
        # Se largura ou altura estiverem faltando, não podemos calcular o custo da cortina
        # Neste caso, o preço pode ser 0 ou podemos levantar um erro, dependendo da regra de negócio.
        # Por enquanto, vamos assumir que não há custo de cortina sem medidas.
        pass

    # 5. Acesse o dicionário price_table para encontrar o preco_m2
    preco_m2_item = 0.0
    min_m2_item = 0.0
    tipo_instalacao_item = None

    if tipo_cortina and tipo_cortina in price_table.get('Tipo de Cortina', {}):
        if subtipo_cortina and subtipo_cortina in price_table['Tipo de Cortina'][tipo_cortina]:
            if tipo_cortina == 'Persiana' and material_cortina and material_cortina in price_table['Tipo de Cortina'][tipo_cortina][subtipo_cortina]:
                item_data = price_table['Tipo de Cortina'][tipo_cortina][subtipo_cortina][material_cortina]
            elif tipo_cortina != 'Persiana' and material_cortina is None:
                item_data = price_table['Tipo de Cortina'][tipo_cortina][subtipo_cortina]
            else:
                item_data = None # Mismatch or missing material for non-persiana
        else:
            item_data = None # Mismatch or missing subtipo
    else:
        item_data = None # Mismatch or missing tipo_cortina

    if item_data:
        preco_m2_item = item_data.get('preco_m2', 0.0)
        min_m2_item = item_data.get('min_m2', 0.0)
        tipo_instalacao_item = item_data.get('tipo_instalacao')

    # 6. Multiplique o preco_m2 pela área e pela quantidade
    effective_area_per_item = max(area_m2, min_m2_item) if min_m2_item > 0 else area_m2
    total_price += (preco_m2_item * effective_area_per_item * quantidade)

    # 7. Se solicita_instalacao for True, adicione o custo de instalação
    if solicita_instalacao and tipo_instalacao_item:
        if tipo_instalacao_item == 'simples':
            total_price += price_table.get('Servicos_Adicionais', {}).get('Instalacao_Padrao', {}).get('custo_fixo', 0.0)
        elif tipo_instalacao_item == 'complexa':
            total_price += price_table.get('Servicos_Adicionais', {}).get('Instalacao_Complexa', {}).get('custo_fixo', 0.0)

    # Outros serviços adicionais podem ser adicionados aqui se o LLM extraí-los
    # Exemplo: Trilho_Suisso ou Var_o_Simples, se o LLM extraísse essa intenção.
    # Por simplicidade, vamos pular por enquanto e focar nos requisitos da subtask.

    # 8. Considere aplicar descontos
    descontos = price_table.get('Descontos', {})

    # Desconto por quantidade/área (ex: Acima de 5m2)
    if 'Quantidade_Acima_5m2' in descontos and (effective_area_per_item * quantidade) > 5.0:
        total_price *= (1 - descontos['Quantidade_Acima_5m2'].get('percentual', 0.0))

    # Desconto para pagamento à vista
    if condicao_pagamento and condicao_pagamento.lower() == 'à vista':
        if 'Pagamento_Vista' in descontos:
            total_price *= (1 - descontos['Pagamento_Vista'].get('percentual', 0.0))

    # 9. Retorne o preco_total calculado e atualize order_details
    order_details['preco_total'] = round(total_price, 2)
    return order_details


print("Função 'calculate_price' definida com sucesso.")

# Exemplo de uso (opcional, para testar a função)
# Suponha que qualify_order_with_llm retornou algo assim:
# sample_order_details = {
#     'cliente_nome': 'João',
#     'cliente_telefone': '(11) 98765-4321',
#     'tipo_cortina': 'Voil',
#     'subtipo_cortina': 'Cor_Unica',
#     'largura': 2.5,
#     'altura': 1.8,
#     'quantidade': 2,
#     'solicita_instalacao': True,
#     'condicao_pagamento': 'à vista',
#     'status_pedido': 'QUALIFICADO'
# }

# updated_order = calculate_price(sample_order_details.copy(), tabela_precos)
# print(f"\nPedido com preço calculado: {updated_order}")
````

## Lógica de Agendamento de Visitas

### Subtask:
Implementar a lógica para processar a solicitação de visita do cliente (identificada pelo LLM). Isso inclui confirmar a disponibilidade, agendar uma `data_visita_agendada` e registrar o `endereco_visita`. Esta informação será usada para atualização no SQLite e Google Sheets.


**Reasoning**:
I need to implement the `schedule_visit` function according to the subtask instructions. This function will check if a visit is requested, simulate scheduling a visit date, record the address, and update the order status accordingly.



````python
from datetime import datetime, timedelta

def schedule_visit(order_details):
    """
    Processa a solicitação de visita do cliente, agenda uma data e registra o endereço.

    Args:
        order_details (dict): Dicionário contendo as informações do pedido extraídas pelo LLM.

    Returns:
        dict: O dicionário order_details atualizado com informações da visita.
    """
    if order_details.get('solicita_visita'):
        # Lógica simples para agendar a visita: 5 dias úteis a partir de hoje
        today = datetime.now()
        # Simula dias úteis, pulando fins de semana. Para maior precisão, seria necessário um calendário.
        scheduled_date = today + timedelta(days=7) # Por simplicidade, adiciona 7 dias corridos.
        order_details['data_visita_agendada'] = scheduled_date.strftime('%Y-%m-%d')

        # Registra o endereço da visita
        endereco = order_details.get('endereco_visita')
        if endereco:
            order_details['endereco_visita'] = endereco
        else:
            # Se o endereço não foi fornecido, adiciona uma observação
            order_details['endereco_visita'] = None # Garante que o campo existe
            current_observation = order_details.get('observacao', '')
            order_details['observacao'] = f"{current_observation} Endereço para visita não fornecido, precisa ser confirmado.\n".strip()

        # Atualiza o status do pedido
        order_details['status_pedido'] = 'VISITA AGENDADA'
    else:
        # Se nenhuma visita for solicitada, garante que os campos de visita estão vazios ou não definidos
        order_details['data_visita_agendada'] = None
        order_details['endereco_visita'] = None

    return order_details

print("Função 'schedule_visit' definida com sucesso.")

# Exemplo de uso (opcional, para testar a função)
# sample_order_details_visit = {
#     'cliente_nome': 'Maria',
#     'cliente_telefone': '(21) 99876-5432',
#     'solicita_visita': True,
#     'data_preferencial_visita': 'próxima semana',
#     'endereco_visita': 'Avenida Principal, 456, Centro, Rio de Janeiro',
#     'status_pedido': 'INCOMPLETO'
# }

# updated_order_visit = schedule_visit(sample_order_details_visit.copy())
# print(f"\nPedido com visita agendada: {updated_order_visit}")

# sample_order_details_no_visit = {
#     'cliente_nome': 'Carlos',
#     'cliente_telefone': '(11) 91234-5678',
#     'solicita_visita': False,
#     'status_pedido': 'QUALIFICADO'
# }
# updated_order_no_visit = schedule_visit(sample_order_details_no_visit.copy())
# print(f"\nPedido sem visita: {updated_order_no_visit}")
````

````python
import sqlite3
from datetime import datetime, timedelta

def save_order_to_sqlite(order_details):
    """
    Salva os detalhes de um pedido na tabela 'pedidos_cortinas' do banco de dados SQLite.

    Args:
        order_details (dict): Dicionário contendo as informações do pedido (qualificadas pelo LLM e calculadas).
    """
    conn = None
    try:
        conn = sqlite3.connect('pedidos_tati.db')
        cursor = conn.cursor()

        # 1. Gerar timestamp atual
        current_timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

        # 2. Calcular data de entrega prevista (7 dias após o timestamp)
        delivery_date = datetime.now() + timedelta(days=7)
        data_entrega_prevista = delivery_date.strftime('%Y-%m-%d')

        # 3. Extrair e formatar campos do order_details
        cliente_nome = order_details.get('cliente_nome', 'N/A')
        cliente_telefone = order_details.get('cliente_telefone', 'N/A')
        tipo_cortina = order_details.get('tipo_cortina', 'N/A')
        subtipo_cortina = order_details.get('subtipo_cortina')
        material_cortina = order_details.get('material_cortina')
        largura = order_details.get('largura', 0.0)
        altura = order_details.get('altura', 0.0)
        quantidade = order_details.get('quantidade', 0)
        preco_total = order_details.get('preco_total', 0.0)
        status_pedido = order_details.get('status_pedido', 'PENDENTE')
        solicita_visita = 1 if order_details.get('solicita_visita', False) else 0 # Convert boolean to integer
        data_visita_agendada = order_details.get('data_visita_agendada')
        endereco_visita = order_details.get('endereco_visita')

        # 4. Construir a instrução SQL INSERT
        insert_query = '''
        INSERT INTO pedidos_cortinas (
            timestamp, cliente_nome, cliente_telefone, tipo_cortina, subtipo_cortina, material_cortina,
            largura, altura, quantidade, preco_total, status_pedido, solicita_visita,
            data_visita_agendada, endereco_visita, data_entrega_prevista
        ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?);
        '''

        # 5. Executar a instrução SQL
        cursor.execute(insert_query, (
            current_timestamp, cliente_nome, cliente_telefone, tipo_cortina, subtipo_cortina, material_cortina,
            largura, altura, quantidade, preco_total, status_pedido, solicita_visita,
            data_visita_agendada, endereco_visita, data_entrega_prevista
        ))

        # 6. Confirmar as alterações
        conn.commit()
        print(f"Pedido para '{cliente_nome}' salvo com sucesso no SQLite (ID: {cursor.lastrowid}).")

    except sqlite3.Error as e:
        print(f"Erro ao salvar pedido no SQLite: {e}")
    finally:
        # 7. Fechar a conexão
        if conn:
            conn.close()

print("Função 'save_order_to_sqlite' definida com sucesso.")

# Exemplo de uso (opcional para teste):
# sample_qualified_order = {
#     'cliente_nome': 'Ana Silva',
#     'cliente_telefone': '(11) 99887-7665',
#     'tipo_cortina': 'Voil',
#     'subtipo_cortina': 'Cor_Unica',
#     'largura': 2.0,
#     'altura': 1.5,
#     'quantidade': 1,
#     'solicita_instalacao': True,
#     'solicita_visita': False,
#     'data_preferencial_visita': None,
#     'endereco_visita': None,
#     'condicao_pagamento': 'boleto',
#     'status_pedido': 'QUALIFICADO',
#     'preco_total': 180.00 # Supondo que já foi calculado
# }
# save_order_to_sqlite(sample_qualified_order)
````

````python
def update_google_sheet(order_details, spreadsheet_name):
    """
    Conecta-se à Google Sheet especificada e adiciona uma nova linha com os dados do pedido.

    Args:
        order_details (dict): Dicionário contendo as informações do pedido (qualificadas, calculadas e agendadas).
        spreadsheet_name (str): O nome da planilha Google Sheets.
    """
    # Certifique-se de que 'gc' (instância autenticada do gspread) está disponível globalmente
    # ou passe-o como um argumento se preferir um escopo mais restrito.
    global gc
    if 'gc' not in globals():
        print("Erro: A instância 'gc' do gspread não foi autenticada. Por favor, execute o passo de autenticação primeiro.")
        return

    try:
        # 1. Conectar à planilha Google Sheets
        spreadsheet = gc.open(spreadsheet_name)
        # 2. Selecionar a primeira aba (worksheet)
        worksheet = spreadsheet.sheet1

        # 3. Preparar a lista de valores para a nova linha
        # A ordem aqui deve corresponder à ordem das colunas na sua Google Sheet.
        # Adapte esta lista para as colunas exatas da sua planilha.
        row_data = [
            order_details.get('timestamp'),
            order_details.get('cliente_nome', 'N/A'),
            order_details.get('cliente_telefone', 'N/A'),
            order_details.get('tipo_cortina', 'N/A'),
            order_details.get('subtipo_cortina'),
            order_details.get('material_cortina'),
            order_details.get('largura'),
            order_details.get('altura'),
            order_details.get('quantidade'),
            order_details.get('preco_total'),
            order_details.get('status_pedido'),
            order_details.get('solicita_visita'), # Será True/False do Python
            order_details.get('data_visita_agendada'),
            order_details.get('endereco_visita'),
            order_details.get('data_entrega_prevista')
        ]

        # Converter booleanos para strings para melhor compatibilidade com planilhas
        row_data = [str(item) if isinstance(item, bool) else item for item in row_data]
        # Substituir None por string vazia para evitar 'None' literal na planilha
        row_data = [item if item is not None else '' for item in row_data]

        # 4. Adicionar a nova linha à planilha
        worksheet.append_row(row_data)
        print(f"Pedido para '{order_details.get('cliente_nome', 'N/A')}' adicionado com sucesso à planilha '{spreadsheet_name}'.")

    except gspread.exceptions.SpreadsheetNotFound:
        print(f"Erro: Planilha '{spreadsheet_name}' não encontrada. Verifique o nome e se a planilha foi compartilhada com a conta de serviço.")
    except gspread.exceptions.APIError as e:
        print(f"Erro na API do Google Sheets: {e}. Verifique as permissões da conta de serviço.")
    except Exception as e:
        print(f"Erro ao atualizar a Google Sheet: {e}")

print("Função 'update_google_sheet' definida com sucesso.")

# Exemplo de uso (com dados fictícios para fins de demonstração):
# from datetime import datetime, timedelta
# sample_data_for_sheet = {
#     'timestamp': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
#     'cliente_nome': 'Cliente Teste',
#     'cliente_telefone': '(99) 99999-9999',
#     'tipo_cortina': 'Voil',
#     'subtipo_cortina': 'Estampado',
#     'material_cortina': None,
#     'largura': 1.8,
#     'altura': 2.2,
#     'quantidade': 1,
#     'preco_total': 187.00,
#     'status_pedido': 'QUALIFICADO',
#     'solicita_visita': False,
#     'data_visita_agendada': None,
#     'endereco_visita': None,
#     'data_entrega_prevista': (datetime.now() + timedelta(days=7)).strftime('%Y-%m-%d')
# }

# # Substitua 'Nome da Sua Planilha' pelo nome da sua planilha de teste
# # update_google_sheet(sample_data_for_sheet, 'Nome da Sua Planilha')
````
